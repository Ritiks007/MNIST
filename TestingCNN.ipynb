```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
from tensorflow.keras.models import load_model
```

```python
img = cv2.imread('6.jpeg')
```

```python
cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(img)
```

```python
img2 = cv2.resize(img,(28,28),interpolation=cv2.INTER_AREA)
```

```python
gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
```

```python
ret,thresh1 = cv2.threshold(gray,132,255,cv2.THRESH_BINARY_INV)
```

```python
plt.imshow(thresh1,cmap='Greys')
```

```python
bl = cv2.GaussianBlur(thresh1,(3,3),0)
```

```python
plt.imshow(bl,cmap='Greys')
```

```python
dil2 = cv2.erode(bl,(3,3))
```

```python
plt.imshow(dil2,cmap='Greys')
```

```python
dil3 = cv2.dilate(dil2,(3,3))
```

```python
plt.imshow(dil3,cmap='Greys')
```

```python
test_img = bl.reshape(1,28,28,1)
```

```python
test_img.shape
```

```python
test_img = test_img.astype('float32')
test_img /= 255
```

```python
model = load_model('cnn_model.h5')
```

```python
pred = model.predict(test_img)
```

```python
print(np.around(pred,4))
print(pred.argmax())
```
